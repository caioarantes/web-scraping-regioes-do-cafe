```python
import pandas as pd
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html = urlopen('https://brazilcoffeenation.com.br/region/list')
soup = BeautifulSoup(html, 'html.parser')
txt = soup.prettify()
txt
```

```python
#note a regiões e o id das subpáginas
re.findall('href="/region/show/id/+[0-9]*">\n           +.*', txt)
```

```python
id_list = re.findall('href="/region/show/id/+([0-9]*)">\n           +.*', txt)
id_list #id das subpaginas das regiões
```

```python
region_list = re.findall('href="/region/show/id/+[0-9]*">\n           +(.*)', txt)
region_list #regiões respectivas aos ids
```

```python
# teste antes do loop para todas as regiões

i = 1 # id teste

html = urlopen('https://brazilcoffeenation.com.br/region/show/id/'+id_list[i])
soup = BeautifulSoup(html, 'html.parser')
txt = soup.prettify()
txt
```

```python
#note o padrão regex para extrair as cidades e UF
re.findall('<div>\n           +(.* - [A-Z]*)', txt)
```

```python
#cidades para cada região em suas respectivas subpáginas em loop, concatenar em uma lista adicinando a região
cidades_todas = []
for i in range(len(region_list)):
    html = urlopen('https://brazilcoffeenation.com.br/region/show/id/'+id_list[i])
    soup = BeautifulSoup(html, 'html.parser')
    txt = soup.prettify()
    cidades = re.findall('<div>\n           +(.* - [A-Z]*)', txt)
    for k in range(len(cidades)):
        cidades[k] =region_list[i]+' - '+cidades[k]
    cidades_todas += cidades
cidades_todas
```

```python
#organizar em colunas dataframe para salvar em csv
df = pd.DataFrame(cidades_todas)
df.head()
```

```python
df[['Região', 'Município', 'UF']] = df[0].str.split(' - ', 2, expand=True)
df = df[['Região', 'Município', 'UF']]
df
```

```python
df.to_csv('cidades_regiao_cafe.csv', encoding = 'latin', index = False)
```
